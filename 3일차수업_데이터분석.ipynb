# 파일처리

| 옵션 | 설명 |
|-|-|
| r | 파일을 읽기모드 처리(원본 보호) |
| w | 파일을 쓰기모드 처리(주의: 덮어쓰기) |
| a | 파일을 추가모드 처리(원본 보호) |
| rb | 파일을 바이트 + 읽기모드 처리(원본 보호) |
| wb | 파일을 바이트 + 쓰기모드 처리(주의: 덮어쓰기) |
| ab | 파일을 바이트 + 추가모드 처리(원본 보호) |

```python
# 파일명 : 1_text.py

# open()함수를 이용하여 텍스트파일이므로 'r' 읽기전용모드로 파일을 호출한다.
# 바이트문자(컴퓨터만 아는 문자), 텍스트문자(사람이 읽을 수 있는 문자)
# 구분 : 메모장으로 열리느냐, 열리지 않느냐
# 메모장으로 열린다 -> read, write, append
# 메모장으로 안열린다 -> rb(binary read), wb(binary write), ab(binary append)

filereader = open('./0711/TEXT/Baby_Shark.txt', 'r', encoding='utf-8')

for row in filereader:
    print(row.strip())

filereader.close()
```

# with 구문을 이용한 파일 호출

일반적으로 open() 함수로 작업을 하면 반드시 close() 함수를 사용하여 해당 객체를 반드시 닫아주어야 한다. 그렇지 않으면 파일 오류가 발생할 수 있다.

with 구문은 파일을 읽고 명령 처리가 끝나면 자동으로 close() 함수를 작동시킨다. 그러므로 close() 함수를 사용하지 않아도 된다.

```python
%%writefile d:/weekend/0711/TEXT/2_text.py
# 파일명 : 2_text.py

with open('./0711/TEXT/Baby_Shark.txt', 'r', newline='', encoding='utf-8') as filereader:        
    for row in filereader:
        print(row.strip())
```

# 폴더를 지정하여 특정 파일 호출하기

# 경로 지정

##### 절대경로 : 드라이브명부터 시작한다. 예) d:/python/0711/a.txt

##### 상대경로 : 현재위치를 의미하는 . (점 하나), 부모위치를 의미하는 .. (점 두개)

                예) 현재 위치에 있는 a.txt --> ./a.txt
                
                예) 부모위치(0711)에 있는 a.txt --> ../a.txt

```python
%%writefile d:/weekend/0711/TEXT/3_text.py

# 파일명 : 3_text.py

import sys
# from glob import glob -> 파일 glob 에서 glob 함수를 가져오는 것
import glob  # 폴더(경로) 지정하는 함수
import os    # 묶어서 한번에 호출

for input_file in glob.glob(os.path.join('./0711/TEXT', '*.txt')):
    with open(input_file, 'r', newline='', encoding='utf-8') as filereader:        
        for row in filereader:
            print(row.strip())
    
```

```python
%%writefile d:/weekend/0711/TEXT/3_text_sys.py

# 파일명 : 3_text_sys.py

import sys
# from glob import glob -> 파일 glob 에서 glob 함수를 가져오는 것
import glob  # 폴더(경로) 지정하는 함수
import os    # 묶어서 한번에 호출

# 파일명이 아닌 파일이 위치한 폴더를 지정해준다.
input_path = sys.argv[1]

# 한글코드명 에러발생시 : encoding='한글코드명' 넣어주기

for input_file in glob.glob(os.path.join(input_path, '*.txt')):
    with open(input_file, 'r', newline='', encoding='utf-8') as filereader:        
        for row in filereader:
            print(row.strip())
    
```

```python
%%writefile  d:/weekend/0711/TEXT/4_text.py

# 파일명 : 4_text.py

import sys 
from glob import glob
import os


for input_file in glob(os.path.join('./0711/TEXT', '*.txt')):
    print(os.path.basename(input_file))
    print('-'*80)
    
    with open(input_file, 'r', newline='', encoding='utf-8') as filereader:
        
        with open('./0711/new_text/merge.txt', 'a', newline='', encoding='utf-8') as filewriter:
            
            for row in filereader:
                filewriter.write(row)
```

```python
%%writefile  d:/weekend/0711/TEXT/4_text_sys.py

# 파일명 : 4_text_sys.py

import sys 
from glob import glob
import os

input_path = sys.argv[1]   # 파일이 있는 폴더 경로 지정
output_file = sys.argv[2]  # 저장할 파일명 지정(input하고 같은 위치가 아닌 곳)


for input_file in glob(os.path.join(input_path, '*.txt')):
    print(os.path.basename(input_file))
    print('-'*80)
    
    with open(input_file, 'r', newline='', encoding='utf-8') as filereader:
        
        with open(output_file, 'a', newline='', encoding='utf-8') as filewriter:
            
            for row in filereader:
                filewriter.write(row)
```

# 텍스트 작성후 저장하기

```python
%%writefile d:/weekend/0711/TEXT/5_text.py

import sys
import os
from glob import glob

my_letters = ['A','B','C','D','E','F','G','H','I','J']

max_index = len(my_letters)

output_file = './0711/new_text/write_save.txt'

with open(output_file, 'w', encoding='utf-8') as filewriter:
    
    for index_value in range(len(my_letters)):
        
        if index_value < (max_index-1):
            filewriter.write(my_letters[index_value] + '\t')
        else:
            filewriter.write(my_letters[index_value] + '\n')
            
print("작성완료")

print(glob(output_file))
```

```python
%%writefile ./0711/PDF/1_pdf.py

import PyPDF2

# PDF 문서를 바이너리 읽기모드로 호출한다.
# 그냥 읽기모드로 가져올 경우 인코딩(언어) 에러 발생
pdfFileObj = open('./0711/PDF/meetingminutes1.pdf', 'rb')

# 지정한 파일을 PyPDF2 함수로 분석
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

# 전체 페이지수 출력
print(pdfReader.numPages)

# 특정 페이지 지정 -> 실습은 첫번재 정보를 가져온다.
pageObj = pdfReader.getPage(0)

# 텍스트문자 추출
pageObj.extractText()

# 서식을 유지하면서 텍스트문자 추출
print(pageObj.extractText())

pdfFileObj.close()
```

# 파일 통합: 회의록1/회의록2

```python
#%writefile ./0711/PDF/4_pdf.py
#파일 통합

import PyPDF2

pdf1FileObj = open('./0711/PDF/meetingminutes1.pdf', 'rb')
pdf2FileObj = open('./0711/PDF/meetingminutes2.pdf', 'rb')

pdf1Reader = PyPDF2.PdfFileReader(pdf1FileObj)
pdf2Reader = PyPDF2.PdfFileReader(pdf2FileObj)

pdfWriter = PyPDF2.PdfFileWriter()

for pageNum in range(pdf1Reader.numPages):
    pageObj = pdf1Reader.getPage(pageNum)
    pdfWriter.addPage(pageObj)
    
for pageNum in range(pdf2Reader.numPages):
    pageObj = pdf2Reader.getPage(pageNum)
    pdfWriter.addPage(pageObj)

# 통합한 파일을 새로운 이름으로 저장
pdfOutputFile = open('./0711/PDF/combined.pdf', 'wb')

pdfWriter.write(pdfOutputFile)
    
pdf1FileObj.close()
pdf2FileObj.close()
pdfOutputFile.close()

print("작업완료")
```

### 페이지 회전 : 90도, 180도, 270도(각각 +,-)

```python
#%%writefile ./0711/PDF/5_pdf.py
# 페이지 회전

import PyPDF2

pdfFileObj = open('./0711/PDF/rotatedPage.pdf', 'rb')

pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
page = pdfReader.getPage(0)
    
# 회전 rotateClockwise(90, 180, 270) 중 선택
page.rotateClockwise(270)   

# --------------------------------------------------------------------

pdfWriter = PyPDF2.PdfFileWriter()
pdfWriter.addPage(page)

resultPdfFile = open('./0711/PDF/rotatedPage_new.pdf', 'wb')

pdfWriter.write(resultPdfFile)

resultPdfFile.close()
pdfFileObj.close()

print("작업완료")
```
